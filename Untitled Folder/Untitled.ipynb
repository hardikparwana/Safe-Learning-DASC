```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
```

```python
u = cp.Variable(1)
a = cp.Parameter(value=0.7)  
b = cp.Parameter(value=0.1)

#(x,a,b)=(0.5,0.4,1) infeasible at 1
#(x,a,b)=(0.5,0.5,1) infeasible at 85
#(x,a,b)=(0.4,0.5,1) success. x=0.6: no solution at 0
#(0.5, 0.9, 0.9), coeff = 0.3
#(0.5, 0.7, 0.2), coeff = 0.7

x = np.array([0.4])
# x >= 0 + t = u
# x <= 1 + 0.3*t only valid till t = 1/0.7. But for some time horizon, it should be possible to keep the constraints

def xdot(x,u):
    return u


def simulate_simple(x0,av,bv,c):
    t = 0
    dt = 0.01
    N = int(1.0/(1.0-c)*100)
    x = x0
    
    a.value = av
    b.value = bv

    for i in range(N):

        #h1
        h1 = x - t # x>=t
        h1_dot = xdot(x,u) - 1

        #h2
        h2 = 1 + c*t - x  #x <= 1 + 0.3t
        h2_dot = c - xdot(x,u)

#         print(h1,h2)
        if h1<0 or h2<0:
            print("************* ERROR ****************")

        #constraints
        const = [h1_dot >= -a*h1]
        const += [h2_dot >= -b*h2]

        objective = cp.Maximize(u)
        problem = cp.Problem(objective,const)
        problem.solve(verbose=False)
        if problem.status != 'optimal':
            print(f"PROBLEM INFEASIBLE at :{i}, a:{a.value}, b:{b.value}")
            return i
            break

        x = x + xdot(x,u.value)*dt
    #     print(f"x_next:{x}, u:{u.value}, h1:{h1}, h1_dot:{h1_dot.value}, -ah1:{-a.value*h1}, h2:{h2}, h2_dot:{h2_dot.value}, -bh2:{-b.value*h2}")
        t += dt
    if i==N-1:
        print("Problem Successful",t)
    return i+1
#     print("time",t)
    
```

```python
e0 = 0.5
nsize = 40
x0 = 0.1
c = 0.3
A = np.linspace(0,5,nsize)
B = np.linspace(0,5,nsize)
horizons_3 = np.zeros((nsize,nsize))
for i in range(nsize):
    for j in range(nsize):
        horizons_3[i,j] = simulate_simple(x0,A[i],B[j],c)
print("DONE")
```

```python
c = 0.3
fig_3, ax_3 = plt.subplots(1)
pos_3 = ax_3.imshow(horizons_3/(1.0/(1-c)),cmap='RdBu', interpolation='bilinear',extent=[0,5,5,0], vmin=0, vmax=100)
fig_3.colorbar(pos_3, ax=ax_3, label='Time Steps to infeasibility')
ax_3.set_ylabel("a")
ax_3.set_xlabel("b")
plt.savefig("paper_example1_3.png")
plt.savefig("paper_example1_3.eps")
```
