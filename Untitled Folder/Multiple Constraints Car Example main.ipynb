### Time varying Safe Set: get more realistic example

```python
u = cp.Variable(1)
a = cp.Parameter(value=0.7)  
b = cp.Parameter(value=0.1)

#(x,a,b)=(0.5,0.4,1) infeasible at 1
#(x,a,b)=(0.5,0.5,1) infeasible at 85
#(x,a,b)=(0.4,0.5,1) success. x=0.6: no solution at 0
#(0.5, 0.9, 0.9), coeff = 0.3
#(0.5, 0.7, 0.2), coeff = 0.7

x = np.array([0.5])
# x >= 0 + t = u
# x <= 1 + 0.3*t only valid till t = 1/0.7. But for some time horizon, it should be possible to keep the constraints

def xdot(x,u):
    return u

t = 0
dt = 0.01
N = 1000

for i in range(N):
    
    #h1
    h1 = x - t # x>=t
    h1_dot = xdot(x,u) - 1

    #h2
    h2 = 1 + 1.1*t - x  #x <= 1 + 0.3t
    h2_dot = 1.1 - xdot(x,u)
    
#     print(h1,h2)

    #constraints
    const = [h1_dot >= -a*h1]
    const += [h2_dot >= -b*h2]

    objective = cp.Maximize(u)
    problem = cp.Problem(objective,const)
    problem.solve(verbose=False)
    if problem.status != 'optimal':
        print("PROBLEM INFEASIBLE at ", i)
        break

    x = x + xdot(x,u.value)*dt
#     print(f"x_next:{x}, u:{u.value}, h1:{h1}, h1_dot:{h1_dot.value}, -ah1:{-a.value*h1}, h2:{h2}, h2_dot:{h2_dot.value}, -bh2:{-b.value*h2}")
    t += dt
if i==N-1:
    print("Problem Successful")
print("time",t)
    
```

### Time Varying Leader Following

```python
u = cp.Variable(1)
a = cp.Parameter(value=0.3)  
b = cp.Parameter(value=0.5)

x = np.array([0.25])
# x >= 0 + t = u
# x <= 1 + 0.3*t only valid till t = 1/0.7. But for some time horizon, it should be possible to keep the constraints

def xdot(x,u):
    return u

def ydot(y,t):
    return 1 + np.sin(20*t)

t = 0
dt = 0.01
N = 1000

y = np.array([1.0])
d_min = 0.2
d_max = 1.5

for i in range(N):
    d = y - x
    
    # min distance
    h1 = d - d_min
    h1_dot = ydot(y,t) - xdot(x,u)
    
    # max distance
    h2 = d_max - d
    h2_dot = -ydot(y,t) + xdot(x,u)
    
    #constraints
    const = [h1_dot >= -a*h1]
    const += [h2_dot >= -b*h2]

    objective = cp.Maximize(u)
    problem = cp.Problem(objective,const)
    problem.solve(verbose=False)
    if problem.status != 'optimal':
        print("PROBLEM INFEASIBLE at ", i)
        break

    # State Update
    x = x + xdot(x,u.value)*dt
    y = y + ydot(y,t)*dt
#     print(f"x_next:{x}, u:{u.value}, h1:{h1}, h1_dot:{h1_dot.value}, -ah1:{-a.value*h1}, h2:{h2}, h2_dot:{h2_dot.value}, -bh2:{-b.value*h2}")
    t += dt
    
if i==N-1:
    print("Problem Successful")
    
```

### Time Varying with different rartes

```python
u = cp.Variable(1)
a = cp.Parameter(value=0.3) #0.2 
b = cp.Parameter(value=0.9) #0.9

x = np.array([-0.4])
# x >= 0 + t = u
# x <= 1 + 0.3*t only valid till t = 1/0.7. But for some time horizon, it should be possible to keep the constraints

coeff = 0.9
E0 = 1.0
D1 = 2
D2 = 2

def xdot(x,u):
    return u

def c1(t):
#     return t
    return t - E0*np.exp( -t/D1 )

def c2(t):
#     return 1 + coeff*t
    return t + 2*E0*np.exp( -t/D2 )

def c1dot(t):
#     return 1
    return 1 + E0/D1*np.exp( -t/D1 )

def c2dot(t):
#     return coeff
    return 1 - 2*E0/D2*np.exp( -t/D2 )

t = 0
dt = 0.01
N = 1000
c1_value = c1(0)
c2_value = c2(0)

for i in range(N):
    
    #h1
    h1 = x - c1(t) # x>=t
#     h1 = x - c1_value
    h1_dot = xdot(x,u) - c1dot(t)

    #h2
    h2 = c2(t) - x  #x <= 1 + 0.3t
#     h2 = c2_value - x  #x <= 1 + 0.3t
    h2_dot = c2dot(t) - xdot(x,u)
    
    if h1<0 or h2<0:
        print(h1,h2)
        print("Safety Violated", i)
        break

    #constraints
    const = [h1_dot >= -a*h1]
    const += [h2_dot >= -b*h2]

    objective = cp.Maximize(u)
    problem = cp.Problem(objective,const)
    problem.solve(verbose=False)
    if problem.status != 'optimal':
        print("PROBLEM INFEASIBLE at ", i)
        break

    x = x + xdot(x,u.value)*dt
    c1_value = c1_value + c1dot(t)*dt
    c2_value = c2_value + c1dot(t)*dt
#     print(f"x_next:{x}, u:{u.value}, h1:{h1}, h1_dot:{h1_dot.value}, -ah1:{-a.value*h1}, h2:{h2}, h2_dot:{h2_dot.value}, -bh2:{-b.value*h2}")
    t += dt
if i==N-1:
    print("Problem Successful")
```

### Time Varying with Different rates discrete time

```python
u = cp.Variable(1)
a = cp.Parameter(value=0.9)  
b = cp.Parameter(value=0.5)
x = np.array([-0.4])

coeff = 0.9
# E0 = 0.5
# D1 = 1
# D2 = 0.4

def xdot(x,u):
    return u

def c1(t,E0,D1):
#     return t
    return t - E0*np.exp( -D1*t )

def c2(t,E0,D2):
#     return 1 + coeff*t
    return t + 2*E0*np.exp( -D2*t )

def c1dot(t,E0,D1):
#     return 1
    return 1 + E0/D1*np.exp( -t/D1 )

def c2dot(t,E0,D2):
#     return coeff
    return 1 - 2*E0/D2*np.exp( -t/D2 )

t = 0; dt = 0.01; N = 1000; h1p = []; h2p = []; up = []



    
def simulate(x0,e0,d1,d2):
    x = x0
    t = 0
    for i in range(N):
        #h1
        print(h1)
        h1 = x - c1(t,e0,d1) # x>=t
        h1_ = h1 + (xdot(x,u) - c1dot(t,e0,d1))*dt

        #h2
        h2 = c2(t,e0,d2) - x  #x <= 1 + 0.3t
        h2_ = h2 + (c2dot(t,e0,d2) - xdot(x,u))*dt

        if h1<0 or h2<0:
            print(h1,h2)
            print("Safety Violated", i)
            print(x,c1(t,e0,d1),c2(t,e0,d2))
            break

        #constraints
        const = [h1_ >= (1-a)*h1]
        const += [h2_ >= (1-b)*h2]

        objective = cp.Maximize(u)
        problem = cp.Problem(objective,const)
        problem.solve(verbose=False)
        if problem.status != 'optimal':
            print("PROBLEM INFEASIBLE at ", i)
            return i
            break
        h1p.append(h1); h2p.append(h2);  up.append(u.value)
        x = x + xdot(x,u.value)*dt
    #     print(f"x_next:{x}, u:{u.value}, h1:{h1}, h1_dot:{h1_dot.value}, -ah1:{-a.value*h1}, h2:{h2}, h2_dot:{h2_dot.value}, -bh2:{-b.value*h2}")
        t += dt
    if i==N-1:
        print("Problem Successful")
        return i+1
```

### Plot simple case

```python
u = cp.Variable(1)
a = cp.Parameter(value=0.7)  
b = cp.Parameter(value=0.1)

#(x,a,b)=(0.5,0.4,1) infeasible at 1
#(x,a,b)=(0.5,0.5,1) infeasible at 85
#(x,a,b)=(0.4,0.5,1) success. x=0.6: no solution at 0
#(0.5, 0.9, 0.9), coeff = 0.3
#(0.5, 0.7, 0.2), coeff = 0.7

x = np.array([0.4])
# x >= 0 + t = u
# x <= 1 + 0.3*t only valid till t = 1/0.7. But for some time horizon, it should be possible to keep the constraints

def xdot(x,u):
    return u


def simulate_simple(x0,av,bv,c):
    t = 0
    dt = 0.01
    N = int(1.0/(1.0-c)*100)
    x = x0
    
    a.value = av
    b.value = bv

    for i in range(N):

        #h1
        h1 = x - t # x>=t
        h1_dot = xdot(x,u) - 1

        #h2
        h2 = 1 + c*t - x  #x <= 1 + 0.3t
        h2_dot = c - xdot(x,u)

#         print(h1,h2)
        if h1<0 or h2<0:
            print("************* ERROR ****************")

        #constraints
        const = [h1_dot >= -a*h1]
        const += [h2_dot >= -b*h2]

        objective = cp.Maximize(u)
        problem = cp.Problem(objective,const)
        problem.solve(verbose=False)
        if problem.status != 'optimal':
            print(f"PROBLEM INFEASIBLE at :{i}, a:{a.value}, b:{b.value}")
            return i
            break

        x = x + xdot(x,u.value)*dt
    #     print(f"x_next:{x}, u:{u.value}, h1:{h1}, h1_dot:{h1_dot.value}, -ah1:{-a.value*h1}, h2:{h2}, h2_dot:{h2_dot.value}, -bh2:{-b.value*h2}")
        t += dt
    if i==N-1:
        print("Problem Successful",t)
    return i+1
#     print("time",t)
    
```
